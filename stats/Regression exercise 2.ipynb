## Linear Regression

```r
tips <- read.csv("../data/tips.csv")
head(tips)
```

```r
summary(tips)
```

```r
hist(tips$tip)
```

```r
mod1 <- lm(tip ~. , data = tips)
summary(mod1)
```

```r
vif(mod1)
```

```r
Table2(mod1)
```

```r
Table2.forestplot(mod1)
```

```r
pred <- predict(mod1, newdata = tips )
hist(pred)
```

```r
scatter.smooth(pred, tips$tip)
```
